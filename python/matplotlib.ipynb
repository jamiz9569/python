# matplotlib graphs


# 1- line graph 

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

x = np.arange(0,10,1)
y = 2 * x + 5

plt.plot(x,y)
plt.show()
```

```python
# customise the line graph 
plt.plot(x,y, color='red', linestyle='--', marker='o') # set the line color, style and marker
plt.title('Line Graph')                     # set the title of the graph
plt.xlabel('X-axis')                        # set the label for x-axis
plt.ylabel('Y-axis')                        # set the label for y-axis
plt.grid()                 # add grid to the graph
plt.legend(['y = 2x + 5']) # add legend to the graph
plt.Figure(figsize=(10,5)) # set the figure size
plt.show()
```

```python
# customise the line graph with multiple lines
y2 = 3 * x + 2
plt.subplot(1,1,1) # create a subplot with 1 row and 1 column
plt.plot(x,y, color='red', linestyle='--', marker='o', label='y = 2x + 5') 
# set the line color, style and marker
plt.subplot(1,1,1) # create a subplot with 1 row and 1 column
plt.plot(x,y2, color='blue', linestyle='-', marker='s', label='y = 3x + 2') 
# set the line color, style and marker
plt.title('Line Graph with Multiple Lines') # set the title of the graph
plt.xlabel('X-axis')                        # set the label for x-axis
plt.ylabel('Y-axis')                        # set the label for y-axis
plt.grid()                 # add grid to the graph
plt.legend() # add legend to the graph
plt.Figure(figsize=(10,5)) # set the figure size
plt.show()
```

# 2- bar plot

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

data = {'apple': 10, 'banana': 15, 'orange': 7, 'grape': 12}

plt.bar(data.keys(), data.values())
plt.xlabel('Fruit')
plt.ylabel('Quantity')
plt.title('Fruit Sales')
plt.show()

plt.barh(data.keys(), data.values())
plt.xlabel('Quantity')
plt.ylabel('Fruit')
plt.title('Fruit Sales')
plt.show()
```


# 3- scatter plot

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

data = {'apple': 10, 'banana': 15, 'orange': 7, 'grape': 12}

plt.scatter(data.keys(), data.values())
plt.xlabel('Fruit')
plt.ylabel('Quantity')
plt.title('Fruit Sales')
plt.show()
```

# 4- histogram

```python
numbers = np.random.rand(100)
plt.hist(numbers, bins=10)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Random Numbers')
plt.show()
```

# 5- box plot and violin plot

```python
total = [10,50,43,15]
order = [10,40,20,61]
discount = [23,42,45,12]
plt.boxplot([total, order, discount], labels=['Total', 'Order', 'Discount'])
plt.ylabel('Value')
plt.title('Boxplot of Sales Data')
plt.show()
```

```python
plt.violinplot([total, order, discount], showmeans=True, showmedians=True)
plt.xticks([1, 2, 3], ['Total', 'Order', 'Discount'])
plt.ylabel('Value')
plt.title('Violin Plot of Sales Data')
plt.show()
```

# 6- piechart and doughnut chart


```python
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is circular.
plt.title('Pie Chart of Categories')
plt.show()

```

```python
groups = ['Group A', 'Group B', 'Group C']
group_size = [20, 35, 30]
size_centre =[5]
pie1 = plt.pie(group_size, labels=groups, autopct='%1.1f%%', startangle=140)
pie2 = plt.pie(size_centre, labels=['Centre'], autopct='%1.1f%%', startangle=140, radius=0.5)
plt.axis('equal')
plt.title('Pie Chart with Centre')
plt.show()
```
