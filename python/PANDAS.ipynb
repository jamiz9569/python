# Merge

Merge in Pandas

Used to combine DataFrames based on a common column.

Syntax
pd.merge(df1, df2, on="column_name")

```python
import pandas as pd

df1 = pd.DataFrame({
    "ID": [1, 2, 3],
    "Name": ["Ali", "Sara", "John"]
})

df2 = pd.DataFrame({
    "ID": [1, 2, 3],
    "Marks": [85, 90, 88]
})

result = pd.merge(df1, df2, on="ID")

print(result)
```

# Types of Merge

```python
    #Inner Merge

#Common values only.

m1 =pd.merge(df1, df2, on="ID", how="inner")
m1

```

```python
    #Left Merge

#All rows from left DataFrame.

m2=pd.merge(df1, df2, on="ID", how="left")
m2

```

```python
    #Right Merge

#All rows from right DataFrame.

m3=pd.merge(df1, df2, on="ID", how="right")
m3

```

```python
    #Outer Merge

#All rows from both DataFrames.

m4=pd.merge(df1, df2, on="ID", how="outer")
m4
```

# joins in pandas

join() is used to combine DataFrames using indexes.

It is similar to merge(), but simpler for index-based joining. Because pandas needed another way to combine tables. Naturally.

Syntax
df1.join(df2)

```python
import pandas as pd

df1 = pd.DataFrame({
    "Name": ["Ali", "Sara", "John"]
})

df2 = pd.DataFrame({
    "Marks": [85, 90, 88]
})

result = df1.join(df2)

print(result)
```

```python
# join using index
```

```python
import pandas as pd

df1 = pd.DataFrame({
    "ID": [1, 2, 3],
    "Name": ["Ali", "Sara", "John"]
}).set_index("ID")

df2 = pd.DataFrame({
    "ID": [1, 2, 3],
    "Marks": [85, 90, 88]
}).set_index("ID")

result = df1.join(df2)

print(result)
```

# types of joins 


Left Join (default)
df1.join(df2, how="left")

Right Join
df1.join(df2, how="right")

Inner Join
df1.join(df2, how="inner")

Outer Join
df1.join(df2, how="outer")

```python
Difference Between Merge and Join
     Merge	     Join
Uses columns	Uses indexes
More flexible	Simpler
 pd.merge()	    df.join()
```

# manipulating the dataset

Data manipulation means:

selecting data
filtering
changing values
adding/removing columns
sorting
grouping

Basically rearranging spreadsheets until the dataset stops looking like evidence from a crime scene.

1. Selecting Columns
Single column
df["Name"]
Multiple columns
df[["Name", "Marks"]]


2. Selecting Rows
Using loc[] (label based)
df.loc[0]
Using iloc[] (index based)
df.iloc[0]


3. Filtering Data
Single condition
df[df["Marks"] > 80]
Multiple conditions
df[(df["Marks"] > 80) & (df["Age"] > 20)]


Use:
& → AND
| → OR


4. Adding New Column
df["Bonus"] = 5
Column using calculation
df["Total"] = df["Marks"] + 10


5. Updating Values
df.loc[0, "Marks"] = 95



6. Deleting Columns
df.drop("Age", axis=1)



7. Deleting Rows
df.drop(0, axis=0)



8. Sorting Data
Ascending order
df.sort_values("Marks")
Descending order
df.sort_values("Marks", ascending=False)



9. Grouping Data
Average marks by department
df.groupby("Department")["Marks"].mean()



Other functions:

sum()
max()
min()
count()


10. Value Counts

Counts frequency of values.

df["Department"].value_counts()


11. Apply Function
df["Marks"] = df["Marks"].apply(lambda x: x + 5)


12. Change Column Names
df.rename(columns={"Marks": "Score"})

```python
import pandas as pd

data = {
    "Name": ["Ali", "Sara", "John"],
    "Marks": [85, 90, 88],
    "Age": [20, 21, 22]
}

df = pd.DataFrame(data)

# Add new column
df["Bonus"] = 5

# Update value
df.loc[0, "Marks"] = 95

# Filter data
result = df[df["Marks"] > 85]

print(result)
```
